## 0. Setup code 

Setup required libraries and parameters for future usage.

```python
import numpy as np 
from numpy.random import normal as norm 
%matplotlib inline 
import matplotlib.pyplot as plt
from matplotlib import ticker 
from matplotlib.ticker import PercentFormatter 
import csv 
params = {'legend.fontsize': 'x-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'large',
         'ytick.labelsize':'large'}
plt.rcParams.update(params)
```

## 1. Initialize Variables 

Initialize required variables, functions and class `Car` 

```python
def avg(l): 
    return float(sum(l)) / len(l) if len(l) != 0 else 0 

def kph2mps(speed): 
    return float(speed) * 1000.0 / 3600.0 

def get_acceleration(speed, time): 
    return speed / time 
```

```python
def auto_poly(): 
    x = np.array([40, 50, 60, 70, 80, 90, 100, 110]) 
    y = np.array([9, 14, 20, 27, 36, 45, 56, 67]) 
    x = x * 1000.0 / 3600.0 
    y = y + 2.5 
    return np.polyfit(x, y, 2) 
    
def manual_poly(): 
    x = np.array([40, 50, 60, 70, 80, 90, 100, 110]) 
    y = np.array([26, 35, 45, 56, 69, 83, 98, 113]) 
    x = x * 1000.0 / 3600.0 
    y = y + 2.5 
    return np.polyfit(x, y, 2) 
    
def safe_auto(speed, poly): 
    return np.polyval(poly, speed)

def safe_manual(speed, poly): 
    return np.polyval(poly, speed)

class Car: 
    def __init__(self, index, comfortableSpeed, acceleration, carType, std, length = 4.4): 
        self.id = index 
        self.pos = 0 
        self.speed = 0 
        self.comfortableSpeed = np.random.normal(comfortableSpeed, comfortableSpeed * std * carType) 
        self.acceleration = np.random.normal(acceleration, acceleration * std * carType) 
        self.length = length 
        self.type = carType # 0 for auto, 1 for manual 
        self.leftFrom = -1 # when leave, change to car id in front 
```

## 2. Double pass simulation 

The simulation consists of two lanes, in which drivers are required to drive at the right lane, and only use the left lane for passing. When the car passes a certain threshold, a car will be added to right lane. When the car have travelled the designated distance, it will be removed from the queue. Cars prefer to drive at its comfortable speed, with the same algorithm as that of the single lane simulation. 

```python
def addCar(queue, threshold, speed, acceleration, std, index): 
    if np.random.uniform() > threshold: 
        queue.insert(0, Car(index, speed, acceleration, 0, 0)) 
    else: 
        queue.insert(0, Car(index, speed, acceleration, 1, std)) 

def move(queue, dt, auto_polynomial, manual_polynomial): 
    for index in range(len(queue)): 
        car = queue[index] 
        if index == len(queue) - 1: 
            if car.speed > car.comfortableSpeed: 
                car.speed -= car.acceleration * dt 
            else: 
                car.speed += car.acceleration * dt 
        else: 
            if (car.type == 0 and queue[index + 1].pos - car.pos < safe_auto(car.speed, auto_polynomial) + car.length) or (car.type == 1 and queue[index + 1].pos - car.pos < safe_manual(car.speed, manual_polynomial) + car.length): 
                car.speed -= car.acceleration * dt 
            elif car.speed < car.comfortableSpeed: 
                car.speed += car.acceleration * dt 
        car.pos += car.speed * dt 

def move1(queue1, dt): 
    for car in queue1: 
        if car.speed < car.comfortableSpeed * 1.2: 
            car.speed += car.acceleration * dt 
        car.pos += car.speed * dt

def checkLast(queue, queue1, distance, end, time): 
    if len(queue) > 0: 
        if queue[-1].pos > distance: 
            index = queue.pop().id 
            end[index] = time 
    if len(queue1) > 0: 
        if queue1[-1].pos > distance: 
            index = queue1.pop().id 
            end[index] = time 

def checkFirst(queue, threshold, start, autoAllow, manualAllow, speed, acceleration, std, time, n, index): 
    if len(queue) == 0: 
        return 0 
    if n > threshold and queue[0].pos > autoAllow: 
        queue.insert(0, Car(index, speed, acceleration, 0, 0)) 
        start[index] = time 
        return 1 
    elif n < threshold and queue[0].pos > manualAllow: 
        queue.insert(0, Car(index, speed, acceleration, 1, std)) 
        start[index] = time 
        return 1 
    return 0 
```

## 3. Takeovers 

Let $c$, $c_1$ and $c_2$ denote three consecutive cars in the slow lane, with $c_2$ being in front of $c_1$, and $c_1$ being in front of $c$. $c$ can make a legal takeover when all of the following conditions suffice: 

1. The current speed of $c$ is less than 0.8 times its comfortable speed 
2. The comfortable speed of $c$ is $20\%$ higher than the comfortable speed of $c_1$. 
3. $c$ has travelled $60$ meters. 
4. The distance between $c_2$ and $c_1$ is greater than the safety distance required for $c$ and $c_1$'s comfortable speed combined.  
5. No other cars are currently passing $c_1$. 

```python
def toFastLane(queue, queue1, auto_polynomial, manual_polynomial): 
    counts = 0 
    for index in range(len(queue) - 3, -1, -1): 
        car = queue[index] 
        car1 = queue[index + 1] 
        car2 = queue[index + 2] 
        cond1 = car.speed / car.comfortableSpeed < 0.8 # test if less than 0.8 
        cond2 = car.comfortableSpeed / car1.comfortableSpeed > 1.2 # test if greater than 1.2 
        cond3 = car.pos > 60 # test if greater than 60 
        cond4 = car2.pos - car1.pos # test if greater than cond5 
        cond5 = safe_auto(car.comfortableSpeed, auto_polynomial) if car.type == 0 else safe_manual(car.comfortableSpeed, manual_polynomial) 
        cond6 = True 
        for cars in queue1: 
            if cars.leftFrom == car1.id: 
                cond6 = False 
        if car1.type == 0: 
            cond5 += safe_auto(car1.comfortableSpeed, auto_polynomial) 
        else: 
            cond5 += safe_manual(car1.comfortableSpeed, manual_polynomial) 
        if cond1 and cond2 and cond3 and cond4 > cond5 + car.length and cond6: 
            car.leftFrom = car1.id 
            queue1.append(queue.pop(index)) 
            counts += 1 
    return counts 
```

## 4. Merge 

Let $c$ denote the car which initiated a takeover, and $c_1$ denote the car taken over. $c$ on the pass lane will merge back to the drive lane when it passes $c_1$ in parallel, and the takeover distance enables an adequate safety distance between $c$ and $c_1$ when $c$ merges backs to the drive lane. For safety reasons, cars are not allowed to pass two cars at the same time. 

```python
def toDriveLane(queue, queue1, auto_polynomial, manual_polynomial): 
    for index1 in range(len(queue1) - 1, -1, -1): 
        car = queue1[index1] 
        insert_index = -1 
        for index2 in range(len(queue)): 
            if queue[index2].id == queue1[index1].leftFrom: 
                car_passed = queue[index2] 
                threshold = car_passed.length 
                if car_passed.type == 0: 
                    threshold += safe_auto(car_passed.speed, auto_polynomial) 
                else: 
                    threshold += safe_manual(car_passed.speed, manual_polynomial) 
                if car.pos - car_passed.pos > threshold: 
                    insert_index = index2 
                break 
        if insert_index != -1: 
            car.leftFrom = -1 
            queue.insert(insert_index + 1, queue1.pop(index1)) 
```

## 4. Single simulation 

The following graph shows the time spent for vehicles to finish the trip under 50% of autonomous vehicles, and 10% of standard deviation of manual driving behavior. 

```python
np.random.seed(2019) 
speed = kph2mps(60) 
acceleration = get_acceleration(speed, 5) 
distance = 1000 
dt = 0.1 
duration = 1200 
auto_polynomial = auto_poly() 
manual_polynomial = manual_poly() 
autoAllow = (acceleration * (safe_auto(kph2mps(60), auto_polynomial) / speed) ** 2) / 2 + 4.4 
manualAllow = (acceleration * (safe_manual(kph2mps(60), manual_polynomial) / speed) ** 2) / 2 + 4.4 
```

```python
std = 0.10 
threshold = 0.5 
np.random.seed(2019) 
start, end, queue, queue1 = {}, {}, [], [] 
start[0] = 0 
time, index = 0, 0 
addCar(queue, threshold, speed, acceleration, std, index) 
index += 1 
n = np.random.uniform() 
overtaken = 0 
while True: 
    move(queue, dt, auto_polynomial, manual_polynomial) 
    move1(queue1, dt) 
    overtaken += toFastLane(queue, queue1, auto_polynomial, manual_polynomial) 
    toDriveLane(queue, queue1, auto_polynomial, manual_polynomial) 
    changed = checkFirst(queue, threshold, start, autoAllow, manualAllow, speed, acceleration, std, time, n, index) 
    if changed == 1: 
        n = np.random.uniform() 
        index += 1 
    checkLast(queue, queue1, distance, end, time) 
    time += 1
    if time > duration / dt: 
        break 
results = {} 
for i in end.keys(): 
    results[i] = float(end[i] - start[i]) / 10 
```

```python
def singleSimulation(data): 
    plt.plot(data, 'ro', color = 'red') 
    plt.title('Car vs. Time - fast pass') 
    plt.xlabel('Car ordered by time of arrival')
    plt.ylabel('Time spent') 
    plt.savefig('single.png', dpi = 200) 
    plt.show()

singleSimulation(list(results.values())) 
```

## 5. Simulation 

Each combination is simulated `100` times, and its statistics are averaged, including: 

1. `passes`: The throughput of the number of cars which finished the travel during the simulation 
2. `means`: The average time took for the cars to finish the travel 
3. `stds`: The standard deviation of the time taken for cars to finish the trip 
4. `mins`: The minimum time taken for the cars to finish the trip 
5. `maxs`: The maximum time taken for the cars to finish the trip 
6. `overtakes`: The number of overtakes occurred during the simulation 

The combinations are as follow: 
$$c = \{(p, s) \mid p \in \{0, 0.1, 0.2, ..., 1.0\} \text{ and } s \in \{0.1, 0.2, ..., 0.20\}\} $$ 

```python
np.random.seed(2019) 
sums = [np.zeros([21, 11]), np.zeros([21, 11]), np.zeros([21, 11]), np.zeros([21, 11]), np.zeros([21, 11]), np.zeros([21, 11])] 
```

```python
for a in range(21): 
    for b in range(11): 
        std = a * 0.01 
        threshold = b * 0.1 
        result = [[], [], [], [], [], []]  
        for c in range(100): 
            start, end, queue, queue1 = {}, {}, [], [] 
            start[0] = 0 
            time, index = 0, 0 
            addCar(queue, threshold, speed, acceleration, std, index) 
            index += 1 
            n = np.random.uniform() 
            overtaken = 0 
            while True: 
                move(queue, dt, auto_polynomial, manual_polynomial) 
                move1(queue1, dt) 
                overtaken += toFastLane(queue, queue1, auto_polynomial, manual_polynomial) 
                toDriveLane(queue, queue1, auto_polynomial, manual_polynomial) 
                changed = checkFirst(queue, threshold, start, autoAllow, manualAllow, speed, acceleration, std, time, n, index) 
                if changed == 1: 
                    n = np.random.uniform() 
                    index += 1 
                checkLast(queue, queue1, distance, end, time) 
                time += 1
                if time > duration / dt: 
                    break 
            result[0].append(len(end.keys())) 
            results = {} 
            for i in end.keys(): 
                results[i] = float(end[i] - start[i]) / 10 
            result[1].append(avg(list(results.values()))) 
            result[2].append(np.std(list(results.values()))) 
            result[3].append(min(list(results.values()))) 
            result[4].append(max(list(results.values()))) 
            result[5].append(overtaken) 
        for m in range(6): 
            sums[m][a, b] = avg(result[m])
    print('Finished %.1f%%' % (float(a + 1) * 100 / 21))
```

## 6. Save as csv 
All results are stored as csv files in the `csv` folder

```python
filenames = ['pass', 'mean', 'std', 'min', 'max', 'overtaken'] 
for i in range(len(sums)): 
    filename = 'csv/' + filenames[i] + '.csv' 
    np.savetxt(filename, sums[i], delimiter = ',')
```

## 7. Plottings 

`plot` function to plot all the following graphs. 

`plotting` displays the results, and the two save_plot functions save the images to `graphs/percentage` and `graphs/standardDeviation`. 

```python
colors = ['#FF0000', '#FF4000', '#FF8000', '#FFBF00', '#FFFF00', '#BFFF00', '#80FF00', 
          '#40FF00', '#00FF00', '#00FF40', '#00FF80', '#00FFBF', '#00FFFF', '#00BFFF', 
          '#0080FF', '#0040FF', '#0000FF', '#4000FF', '#8000FF', '#BF00FF', '#FF00FF'] 
legends1 = ["%.2f" % i for i in np.linspace(0, 0.2, 21)] 
legends2 = ["%d%%" % (i * 100) for i in np.linspace(0, 1, 11)]

def plotting(data, title, legends_loc1, legends_loc2, l_size1 = 12, l_size = 12): 
    xticks1 = ticker.MaxNLocator(12) 
    xticks2 = ticker.MaxNLocator(12) 
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (20, 8)) 
    fig.suptitle(title, fontsize = 18)
    for i in range(len(data)): 
        ax1.plot(np.linspace(0, 1, 11), data[i], label = legends1[i], color = colors[i]) 
    ax1.legend(loc = legends_loc1, ncol = 3, prop = {'size': l_size1}) 
    d_copy = np.transpose(data) 
    for i in range(len(d_copy)): 
        ax2.plot(np.linspace(0, 0.20, 21), d_copy[i], label = legends2[i], color = colors[i * 2]) 
    ax2.legend(loc = legends_loc2, ncol = 4, prop = {'size': l_size}) 
    ax1.xaxis.set_major_locator(xticks1) 
    ax2.xaxis.set_major_locator(xticks2) 
    ax1.set_title('Autonomous Percentage vs. ' + title) 
    ax2.set_title('SDMB vs. ' + title) 
    xvals = ax1.get_xticks() 
    ax1.set_xticklabels(['{:,.0%}'.format(x) for x in xvals]) 
    ax2.set_xlabel('Standard Deviation') 
    ax2.set_ylabel(title) 
    ax1.set_xlabel('Autonomous Percentage') 
    ax1.set_ylabel(title) 
    plt.show()
```

```python
def save_plot(data, title, legend_loc, l_size = 6): 
    fig = plt.figure() 
    plt.gcf().subplots_adjust(bottom=0.15) 
    ax = fig.add_subplot(1, 1, 1) 
    xticks = ticker.MaxNLocator(12) 
    for i in range(len(data)): 
        ax.plot(np.linspace(0, 1, 11), data[i], color = colors[i], label = legends1[i]) 
    filename = title.lower() + '.png' 
    ax.legend(loc = legend_loc, ncol = 3, prop = {'size': l_size})
    ax.xaxis.set_major_locator(xticks) 
    xvals = ax.get_xticks() 
    ax.set_title('Autonomous Percentage vs. ' + title, size = 10) 
    ax.set_xlabel('Autonomous Percentage') 
    ax.set_ylabel(title) 
    ax.set_xticklabels(['{:,.0%}'.format(x) for x in xvals]) 
    fig.savefig('graphs/percentage/' + filename, dpi = 200, bbox_inches = 'tight') 
    plt.close(fig)
```

```python
def save_plot2(data, title, legend_loc, l_size = 6): 
    fig = plt.figure() 
    plt.gcf().subplots_adjust(bottom=0.15) 
    ax = fig.add_subplot(1, 1, 1) 
    xticks = ticker.MaxNLocator(12) 
    d_copy = np.transpose(data) 
    for i in range(len(d_copy)): 
        ax.plot(np.linspace(0, 0.20, 21), d_copy[i], color = colors[i * 2], label = legends2[i]) 
    filename = title.lower() + ' .png' 
    ax.legend(loc = legend_loc, ncol = 4, prop = {'size': l_size}) 
    ax.xaxis.set_major_locator(xticks) 
    xvals = ax.get_xticks() 
    ax.set_title('SDMB vs. ' + title, size = 10) 
    ax.set_xlabel('SDMB') 
    ax.set_ylabel(title) 
    ax.set_xticklabels(['{:,.0%}'.format(x) for x in xvals]) 
    fig.savefig('graphs/standardDeviation/' + filename, dpi = 200, bbox_inches = 'tight')  
    plt.close(fig)
```

```python
def plot(data, title, legend_loc1 = 'upper left', legend_loc2 = 'upper right', l_size1 = 12, l_size2 = 12, s_size = 6): 
    plotting(data, title, legend_loc1, legend_loc2, l_size1, l_size2) 
    save_plot(data, title, legend_loc = legend_loc1, l_size = s_size) 
    save_plot2(data, title, legend_loc = legend_loc2, l_size = s_size)
```
